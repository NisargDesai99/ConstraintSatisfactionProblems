```python
def parse_input(var_file_path, constraints_file_path):
    var_file = open(var_file_path)
    constraint_file = open(constraints_file_path)

    var_file_lines = var_file.readlines()

    variables = {}
    for line in var_file_lines:
        line_split = line.strip().split(':')

        var = line_split[0].strip()
        domain = line_split[1].strip().split(' ')

        variables[var] = domain

    constraint_file_lines = constraint_file.readlines()
    constraints = []
    for line in constraint_file_lines:
        line_split = line.strip().split(' ')
        constraints.append([item for item in line_split])
    
    return (variables, constraints)
```
